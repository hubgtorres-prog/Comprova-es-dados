Operações Basicas no phyton

```python
# somar uma coluna
print(df['total_venda'].sum())
```

```python
# Multiplicar colunas
df['total_venda'] = df['quantidade'] * df['preco_unitario']
print(df['total_venda'])
```

```python
# Media da coluna quantidade
print(df['quantidade'].mean())
```

Filtros

```python
# Filtrar itens da categorias eletrônicos
print(df[df['categoria'] == 'Eletrônicos'])
```

```python
# Filtros compostos ou combinados
print(df[(df['categoria'] == 'Eletrônicos') & (df['quantidade'] > 3)])
```

Agrupamentos

```python
# Media de quandidade vendida por categoria
print(df.groupby('categoria')['quantidade'].mean())
```

```python
# Soma de quandidade vendida por categoria
print(df.groupby('categoria')['quantidade'].sum())
```

```python
# Multiplas estaticas por categoria
print(df.groupby('categoria')['quantidade'].agg(['mean', 'sum','count','max','min']))
```

```python
# 1. Corrigir e converter a coluna preco_unitario
df['preco_unitario'] = (
    df['preco_unitario']
    .astype(str)                                  # garante que é string
    .str.replace('.', '', regex=False)            # remove ponto dos milhares
    .str.replace(',', '.', regex=False)           # troca vírgula decimal por ponto
)

# 2. Converter para float
df['preco_unitario'] = pd.to_numeric(df['preco_unitario'], errors='coerce')

# 3. Converter quantidade, se necessário
df['quantidade'] = pd.to_numeric(df['quantidade'], errors='coerce')

# 4. Multiplicar
df['total_venda'] = df['quantidade'] * df['preco_unitario']

# 5. Verificar tipo final
print(df['total_venda'].dtype)   # deve retornar float64

```

Criar uma coluna nova VIP ou NAO VIP

```python
# Criar uma noa coluna STATUS do Cliente
df['status'] = np.where(df['total_venda'] > 10000, 'VIP', 'NAO VIP')
print(df['status'])
```

```python
# Vendo head do data set
print(df.head())
```

Trabalhando com Datas

```python
# Garantir que as informações de datas estão no formato datetime
df['data_venda'] = pd.to_datetime(df['data_venda'])

#
```

```python
# extratir as informações de data
df['ano'] = df['data_venda'].dt.year
df['mes'] = df['data_venda'].dt.month
df['dia'] = df['data_venda'].dt.day
df['dia_da_semana'] = df['data_venda'].dt.dayofweek
```

Analise de Oportunidades

```python
# qual o total de vendas
print(df['total_venda'].sum())
```

```python
# qual a media de vendas
print(df['total_venda'].mean())
```

```python
# qual categoria vende mais
print(df.groupby('categoria')['total_venda'].sum())
```

```python
# maior venda individual?
print(df['total_venda'].max())
```

```python
# dia da semana com maior venda
print(df.groupby('dia_da_semana')['total_venda'].sum())
```

```python
# Calcular a semana do mês (aproximada)
df['semana_do_mes'] = np.ceil(df['dia'] / 7)

# Venda total por semana do mês
print(df.groupby('semana_do_mes')['total_venda'].sum())
```

Analise final

```python
# grafico de vendas totais por mes e ano
plt.figure(figsize=(12, 6))
df.groupby(['ano', 'mes'])['total_venda'].sum().sort_index().plot(kind='bar')
plt.title('Total de Vendas por Mês e Ano')
plt.xlabel('Ano e Mês')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# graficos de pizzas de um porcentual de vendas por categoria
plt.figure(figsize=(10, 6))
df['categoria'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporção de Vendas por Categoria')
plt.ylabel('')  # Remove default y-label
plt.legend(title='Categorias', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```
